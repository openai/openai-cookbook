# How AI Agents Are Replacing the Slowest Part of Paid Media

The slowest part of paid media isn't strategy — it's execution. Translating a campaign brief into live ads means navigating Google Ads Manager, switching to Meta Business Suite, building ad sets, uploading creatives, writing headline variations, and configuring extensions. That's where the day goes.

This cookbook shows how to use the [Adspirer MCP server](https://www.adspirer.com) with ChatGPT to **create a complete multi-platform ad campaign — Google Search + Meta Ads with 8+ creative variants — in a single conversation**.

| What | Manual (platform UIs) | With ChatGPT + Adspirer MCP |
|------|----------------------|-----------------------------|
| Research keywords + audiences | 30-60 min | 1-2 prompts |
| Create Google Search campaign with extensions | 30-45 min | 1-2 prompts |
| Create Meta campaign with 8 creative variants | 75-150 min | 2 prompts |
| Set up monitoring alerts | 15-20 min | 1 prompt |
| **Total** | **4-8 hours** | **~15 minutes** |

These numbers come from real production data: a DTC brand tested 15 distinct messaging angles with 174 ad operations in 10 days. An agency launched 64 campaigns across 5 formats for clients in Dubai and Delhi from one chat window.

Add Adspirer as a ChatGPT connector:

1. Open ChatGPT → **Settings → Connectors**
2. Click **Add custom connector**
3. Enter:
   - **Name:** Ads MCP
   - **URL:** `https://mcp.adspirer.com/mcp`

4. Complete the OAuth sign-in flow
5. Link your Google Ads and/or Meta Ads accounts

## The Scenario

You're launching ads for a **direct-to-consumer skincare brand** that sells a Vitamin C serum. You want:

- A **Google Search campaign** capturing people actively searching for vitamin C serums
- A **Meta Ads campaign** with multiple creative variants testing different messaging angles
- **Monitoring** to catch problems early

Let's build all of it in one ChatGPT conversation.

---

## Step 1: Check Your Connected Platforms

Before creating anything, verify which ad platforms are linked.

**Prompt:**

```
What ad platforms do I have connected? Show me the account details.
```

ChatGPT calls `get_connections_status`, which returns your connected platforms, account names, and IDs.

**Expected output:**

```
Connected Platforms:
┌─────────────┬──────────────────────────┬────────────────┬─────────┐
│ Platform    │ Account Name             │ Account ID     │ Status  │
├─────────────┼──────────────────────────┼────────────────┼─────────┤
│ Google Ads  │ My Skincare Brand        │ 123-456-7890   │ Active  │
│ Meta Ads    │ Skincare-Ads-Account     │ 129971168...   │ Active  │
└─────────────┴──────────────────────────┴────────────────┴─────────┘
```

> If a platform isn't connected, go to [adspirer.com](https://www.adspirer.com) → Connections to link your ad accounts.

## Step 3: Create a Google Search Campaign

One prompt creates the campaign, ad group, keywords, and ad copy.

**Prompt:**

```
Create a Google Search campaign for my Vitamin C serum brand:

- Campaign name: "Search - Vitamin C Serum - US"
- Budget: $50/day
- Location: United States
- Keywords: "best vitamin c serum", "vitamin c serum for face",
  "buy vitamin c serum", "vitamin c serum for dark spots",
  "vitamin c serum sensitive skin"
- Write 5 headline variations and 3 descriptions
- Bidding: maximize conversions
```

ChatGPT calls `create_search_campaign`. The campaign is created in **PAUSED** status — it won't spend money until you activate it.

**Expected output:**

```
✅ Campaign created (PAUSED):

Campaign: Search - Vitamin C Serum - US
├── Budget: $50.00/day
├── Bidding: Maximize Conversions
├── Location: United States
│
├── Ad Group: Vitamin C Serum
│   ├── Keywords (5):
│   │   ├── "best vitamin c serum" (phrase match)
│   │   ├── "vitamin c serum for face" (phrase match)
│   │   ├── "buy vitamin c serum" (phrase match)
│   │   ├── "vitamin c serum for dark spots" (phrase match)
│   │   └── "vitamin c serum sensitive skin" (phrase match)
│   │
│   └── Ad:
│       ├── Headlines:
│       │   ├── "Best Vitamin C Serum 2026"
│       │   ├── "Brighter Skin in 14 Days"
│       │   ├── "Dermatologist-Recommended"
│       │   ├── "20% Vitamin C + Hyaluronic Acid"
│       │   └── "Free Shipping on All Orders"
│       └── Descriptions:
│           ├── "Clinically proven to reduce dark spots. Pure 20% L-ascorbic acid. Shop now."
│           ├── "Join 50,000+ customers who transformed their skin. 30-day money-back guarantee."
│           └── "Lightweight, fast-absorbing serum for all skin types. Vegan & cruelty-free."
│
└── Status: PAUSED (activate when ready)
```

### Add Ad Extensions

Extensions make your ads more prominent and drive higher CTR.

**Prompt:**

```
Add sitelink extensions for these pages:
- "Shop Serums" → /collections/serums
- "Before & After Results" → /results
- "Our Ingredients" → /ingredients
- "Reviews" → /reviews

Also add callout extensions: "Free Shipping", "30-Day Guarantee",
"Dermatologist Tested", "Vegan & Cruelty-Free"
```

ChatGPT calls `add_sitelinks` and `add_callout_extensions`.

## Step 4: Create Meta Ads with Multiple Creative Variants

This is where the time savings are most dramatic. In Meta's Ads Manager, creating each ad variant means navigating the creative builder, uploading assets, writing copy, and configuring the CTA — 5-10 minutes per variant.

Via conversation, you describe the angles you want to test and they're all created at once.

**Prompt:**

```
Create a Meta image campaign for my Vitamin C serum:

- Campaign name: "Meta - Vitamin C Serum - Prospecting"
- Budget: $30/day
- Objective: Sales
- Targeting: Women 25-45 in the US interested in skincare,
  anti-aging, clean beauty
- Image: use my existing product photo (discover from my assets)

Create 5 ad variants, each with a different messaging angle:

1. SOCIAL PROOF: "Join 50,000+ customers who transformed their skin.
   Visible results in 14 days."

2. RISK REVERSAL: "Not sure if it'll work for you? Try it risk-free
   with our 30-day money-back guarantee."

3. SCIENCE/AUTHORITY: "20% L-Ascorbic Acid — the gold standard
   concentration recommended by dermatologists."

4. PAIN POINT: "Dark spots? Uneven skin tone? Our Vitamin C serum
   fades hyperpigmentation at the cellular level."

5. BEFORE/AFTER: "Real results from real customers. See the
   transformation in just 2 weeks."
```

ChatGPT calls `discover_meta_assets` to find existing images, then `create_meta_image_campaign` with all 5 variants.

**Expected output:**

```
✅ Meta campaign created (PAUSED):

Campaign: Meta - Vitamin C Serum - Prospecting
├── Objective: Sales
├── Budget: $30.00/day
├── Targeting: Women 25-45, US, skincare + anti-aging + clean beauty
│
├── Ad Set: Prospecting - Skincare Enthusiasts
│   ├── Ad 1 (Social Proof): "Join 50,000+ customers..."
│   ├── Ad 2 (Risk Reversal): "Not sure if it'll work for you?..."
│   ├── Ad 3 (Science): "20% L-Ascorbic Acid — the gold standard..."
│   ├── Ad 4 (Pain Point): "Dark spots? Uneven skin tone?..."
│   └── Ad 5 (Before/After): "Real results from real customers..."
│
└── Status: PAUSED
```

### Add Retargeting Variants

People who visited your site but didn't buy need different messaging.

**Prompt:**

```
Add 3 more ad variants for retargeting — people who visited
but didn't purchase:

1. "You were looking at our Vitamin C serum. Still thinking
   about it? Your skin isn't going to wait. Free shipping today."

2. "You did your research. Now try the real thing — 30-day
   guarantee, no risk."

3. "Still on the fence? Here's what 50,000+ customers say about
   their results after just 14 days."
```

ChatGPT calls `add_meta_ad` for each variant.

## Step 5: Set Up Monitoring

Don't wait for problems to find you.

**Prompt:**

```
Set up monitoring for both campaigns:
- Alert me if CPA exceeds $25 on either platform
- Alert me if CTR drops below 1% on Google or 0.5% on Meta
- Send a weekly performance summary every Monday
```

ChatGPT calls `create_monitor` for the alerts and `schedule_brief` for the weekly summary.

## What Comes Next

Your campaigns are created in PAUSED status. Here's the recommended cadence:

| When | Action | Prompt |
|------|--------|--------|
| **Now** | Review and activate in each platform's UI | — |
| **Day 7** | First performance review | "Run a cross-platform performance review for the last 7 days." |
| **Day 14** | Wasted spend audit | "Analyze wasted spend across both platforms. Which keywords or placements are burning budget?" |
| **Day 14** | Creative fatigue check | "Check my Meta ads for creative fatigue. Which variants have declining CTR?" |
| **Day 21** | Optimization round | "Based on 3 weeks of data, pause underperformers and scale winners." |
